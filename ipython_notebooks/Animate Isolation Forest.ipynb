```python
mean = [2.0, -1.0]
cov = [[5, 0], [0, 1.5]]
x, y = np.random.multivariate_normal(mean, cov, 500).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.axis((-8.0, 10.0, -8.0, 4.0))
```

```python
X = np.column_stack((x, y))
angle = np.radians(33.0)
rot_mat = np.array([[np.cos(angle), -np.sin(angle)], 
                    [np.sin(angle),  np.cos(angle)]])
newX = X.dot(rot_mat)
covar = np.cov(newX, rowvar=0)
print covar
evals, evecs = np.linalg.eig(covar)
print evals
print evecs
```

```python
x = newX[:,0]
y = newX[:,1]
plt.plot(x, y, 'x')
plt.axis('equal')
plt.axis((-8.0, 10.0, -8.0, 4.0))
plt.savefig('BiMaxwellian0.png')
```

```python
import matplotlib.mlab as mlab
xg = np.arange(-8.0, 10.0, 0.1)
yg = np.arange(-8.0, 4.0, 0.1)
Xg, Yg = np.meshgrid(xg, yg)

plt.plot(x, y, 'x')
Z = mlab.bivariate_normal(Xg, Yg, covar[0,0], covar[1,1], np.mean(x), np.mean(y), covar[0,1])
plt.contour(Xg,Yg,Z)
plt.axis((-8.0, 10.0, -8.0, 4.0))
plt.savefig('BiMaxwellian1.png')
```

```python
from scipy import stats
m0, m1 = X[:,0], X[:,1]
print np.shape(m0), np.shape(m1)
positions = np.vstack([Xg.ravel(), Yg.ravel()])
values = np.vstack([m0, m1])
kernel = stats.gaussian_kde(values)
print np.shape(kernel(positions))
f = np.reshape(kernel(positions).T, Xg.shape)
plt.plot(x, y, 'x')
levels = np.linspace(np.nanmin(f), np.nanmax(f), 25)
print levels
plt.contour(Xg, Yg, f, levels=levels)
plt.axis((-8.0, 10.0, -8.0, 4.0))
plt.savefig('KDEanomaly.png')
```
